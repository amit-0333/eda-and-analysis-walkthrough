```python
# | Source              | Example                                                                    |
# | ------------------- | -------------------------------------------------------------------------- |
# | Direct CSV URL      | `pd.read_csv('https://example.com/data.csv')`                              |
# | GitHub Raw File     | `pd.read_csv('https://raw.githubusercontent.com/user/repo/main/file.csv')` |
# | Kaggle Download URL | `pd.read_csv('URL')` *(if publicly accessible)*                            |
# | Local File          | `pd.read_csv('/content/file.csv')`                                         |


import pandas as pd
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

req = requests.get(url)
data = StringIO(req.text)

df = pd.read_csv(
    data,
    usecols=['Country', 'Region'],
    nrows=20,
    index_col='Country'
)

print(df.head())
```

# **parse_dates: Converts specified date/time columns from string (object) to datetime64[ns] while reading a CSV file.**

```python
## Handling Dates/timestamps : Used to automatically convert date/timestamp columns from string (object) to datetime (datetime64[ns]) while reading the CSV.

pd.read_csv('/content/transactions.csv',parse_dates=['Timestamp']).info()
```

```python
# | `parse_dates` Value                  | Purpose                                                        | Example                                                                   |
# | ------------------------------------ | -------------------------------------------------------------- | ------------------------------------------------------------------------- |
# | `['Timestamp']`                      | Convert one date column                                        | `pd.read_csv('file.csv', parse_dates=['Timestamp'])`                      |
# | `['Order_Date', 'Ship_Date']`        | Convert multiple date columns                                  | `pd.read_csv('file.csv', parse_dates=['Order_Date', 'Ship_Date'])`        |
# | `[['Year', 'Month', 'Day']]`         | Combine multiple columns into one datetime column              | `pd.read_csv('file.csv', parse_dates=[['Year', 'Month', 'Day']])`         |
# | `{'Date': ['Year', 'Month', 'Day']}` | Combine columns and give the new datetime column a custom name | `pd.read_csv('file.csv', parse_dates={'Date': ['Year', 'Month', 'Day']})` |
# | `['Timestamp', 'Order_Date']`        | Convert several date columns at once                           | `pd.read_csv('file.csv', parse_dates=['Timestamp', 'Order_Date'])`        |
```

# **converters : converters applies a function to every value in the specified column while reading the CSV.**

```python
## converters : converters applies a function to every value in the specified column while reading the CSV.

def status(status):
  if status=='SUCCESS':
    return 1
  return 0


pd.read_csv('/content/transactions.csv',converters={'Status' : status})
```

# **na_values: Specifies additional values that should be treated as missing values (NaN) while reading a CSV file.**

```python
## na_values parameter : na_values tells Pandas which values should be treated as missing values (NaN).

#  na_values specifies additional values that should be interpreted as missing (NaN) while reading a CSV file.

pd.read_csv('/content/transactions.csv',na_values=['NA', 'N/A', '?', '--'])
```

```python
# | `na_values` Value                             | Meaning                              | Example                                                                          |
# | --------------------------------------------- | ------------------------------------ | -------------------------------------------------------------------------------- |
# | `'NA'`                                        | Treat `NA` as missing                | `pd.read_csv('file.csv', na_values='NA')`                                        |
# | `'?'`                                         | Treat `?` as missing                 | `pd.read_csv('file.csv', na_values='?')`                                         |
# | `['NA', '?']`                                 | Multiple missing value indicators    | `pd.read_csv('file.csv', na_values=['NA', '?'])`                                 |
# | `['NA', 'N/A', 'NULL']`                       | Multiple text values as NaN          | `pd.read_csv('file.csv', na_values=['NA', 'N/A', 'NULL'])`                       |
# | `['--', 'Unknown']`                           | Custom missing value indicators      | `pd.read_csv('file.csv', na_values=['--', 'Unknown'])`                           |
# | `{'Status':['Unknown']}`                      | Column-specific missing values       | `pd.read_csv('file.csv', na_values={'Status':['Unknown']})`                      |
# | `{'Amount':['?','NA']}`                       | Different NaN values for a column    | `pd.read_csv('file.csv', na_values={'Amount':['?','NA']})`                       |
# | `{'Status':['Unknown'], 'Amount':['?','NA']}` | Different rules for multiple columns | `pd.read_csv('file.csv', na_values={'Status':['Unknown'], 'Amount':['?','NA']})` |
```

# **chunksize: Reads a large CSV file in smaller batches (chunks) instead of loading the entire dataset into memory at once.**

```python
## Loading a huge dataset in chunks : chunksize reads a large CSV file in smaller batches (chunks) instead of loading the entire dataset into memory at once.

dfs=pd.read_csv('/content/india-districts-census-2011.csv',chunksize=210)
```

```python
for chunk in dfs:
  print(chunk.shape)


# Remember: After iterating through dfs once, it's exhausted. If you need it again, recreate it with pd.read_csv(..., chunksize=210).
```

```python
# | Without `chunksize`        | With `chunksize`            |
# | -------------------------- | --------------------------- |
# | Loads entire file into RAM | Loads small parts at a time |
# | Can cause memory errors    | Memory efficient            |
# | Slow for huge files        | Better for large datasets   |
```

```python
total_pop = 0

for chunk in pd.read_csv(
    '/content/india-districts-census-2011.csv',
    chunksize=210
):
    total_pop += chunk['Population'].sum()

print(total_pop)
```
