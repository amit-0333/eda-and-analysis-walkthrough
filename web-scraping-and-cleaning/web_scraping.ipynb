```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
books = soup.find_all('article', class_='product_pod')

print(len(books))
```

```python
book = books[0]

print(book.prettify())
```

```python
book.find('img')['src']
```

```python
titles = []
prices = []
ratings = []
availability = []

for book in books:

    titles.append(book.h3.a['title'])

    prices.append(
        book.find('p', class_='price_color').text
    )

    ratings.append(
        book.find('p', class_='star-rating')['class'][1]
    )

    availability.append(
        book.find(
            'p',
            class_='instock availability'
        ).text.strip()
    )
```

```python
len(titles)
```

```python
df = pd.DataFrame({
    'Title': titles,
    'Price': prices,
    'Rating': ratings,
    'Availability': availability
})

df.head()
```

# **Fake Job**

```python
from bs4 import BeautifulSoup
import requests

headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(
    'https://realpython.github.io/fake-jobs/',
    headers=headers
)

print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')

print(soup.title)
```

```python
# print(soup.prettify())
```

```python
jobs = soup.find_all("div", class_="card-content")
jobs[0]
```

```python
print("Total jobs:", len(jobs))
```

```python
first_job = jobs[0]

title = first_job.find('h2').text.strip()
company = first_job.find('h3').text.strip()
location = first_job.find('p', class_='location').text.strip()

print(title)
print(company)
print(location)
```

```python
jobs = soup.find_all("div", class_="card-content")

titles = []
companies = []
locations = []

for job in jobs:
    titles.append(job.find("h2").text.strip())
    companies.append(job.find("h3").text.strip())
    locations.append(job.find("p", class_="location").text.strip())

len(titles)
len(companies)
len(locations)
```

```python

df = pd.DataFrame({
    "title": titles,
    "company": companies,
    "location": locations
})

df.head()
```
